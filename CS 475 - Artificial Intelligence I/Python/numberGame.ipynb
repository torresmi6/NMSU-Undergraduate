```python
from games import *

class numberGame(Game):
    
    def __init__(self, initial):
        self.initial = initial

    def actions(self, state):
        actions = []
        if not state:
            for e in range(10):
                actions.append(e)
            return actions
        
        if (state[-1] % 2) == 0:
            for x in range(state[-1]):
                if not is_in(x, list(state)):
                    actions.append(x)
        else:
            if not is_in("3", list(state)):
                actions.append("3")
            if not is_in("6", list(state)):
                actions.append("6")
            if not is_in("9", list(state)):
                actions.append("9")
        if state == 2 or state == 3 or state == 5 or state == 7:
            prime = ["2","3","5","7"]
            for ele in prime:
                if not is_in(ele, list(state)):
                    actions.append(ele)
        return actions
    
    def result(self, state, move):
        state += move
        return state
    
    def utility(self, state, player):
        return state[-1]
    
    def terminal_test(self, state):
        return not self.actions(state)
    
    def to_move(self, state):
        return state.to_move
            
    def play_game(self, *players):
        """Play an n-person, move-alternating game."""
        state = self.initial
        while True:
            for player in players:
                move = player(self, state)
                state = self.result(state, move)
                if self.terminal_test(state):
                    self.display(state)
                    return self.utility(state, self.to_move(self.initial))
    
    def minmax_decision(state, game):
        """Given a state in a game, calculate the best move by searching
        forward all the way to the terminal states. [Figure 5.3]"""
        player = game.to_move(state)

        def max_value(state):
            if game.terminal_test(state):
                return game.utility(state, player)
            v = -np.inf
            for a in game.actions(state):
                v = max(v, min_value(game.result(state, a)))
            return v

        def min_value(state):
            if game.terminal_test(state):
                return game.utility(state, player)
            v = np.inf
            for a in game.actions(state):
                v = min(v, max_value(game.result(state, a)))
            return v

        # Body of minmax_decision:
        return max(game.actions(state), key=lambda a: min_value(game.result(state, a)))

def main():
    initial = []
    game = numberGame(initial)
    print(minmax_decision(initial, game))
    
main()
```
